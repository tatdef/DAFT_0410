### 2. Use crosstabs to show the number of moovies per country

```python
moovies= pd.read_csv("movie_metadata.csv")
```

```python
moovies.describe(include='O')
```

```python
pd.crosstab(index=[moovies['country']], columns='country', values='country', aggfunc=len)
```

### 3. Create a crosstab to show the number of films per director and genre splitted by country

```python
pd.crosstab([moovies["genres"],moovies["director_name"]], [moovies["country"]], margins=True)
```

### 4. Average duration of filmes grouped by color

```python
moovies.columns
```

```python
moovies['color'].value_counts()
```

```python
pd.crosstab(moovies['color'], moovies['color'], values= moovies['duration'], aggfunc='mean', margins=True, margins_name='Total')[["Total"]].reset_index()
```
