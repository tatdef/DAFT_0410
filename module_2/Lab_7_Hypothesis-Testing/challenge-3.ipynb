# Challenge 2 - Hypothesis Tests

In this section of the lab, we will test whether the hourly wage of all hourly workers is significantly different from $30/hr. Import the correct one sample test function from scipy and perform the hypothesis test for a 95% two sided confidence interval.

```python
# assuming that the salaries have a normal distribution 
import scipy
from scipy import stats 
hourly_workers = list(salaries.loc[salaries['Salary or Hourly']=='Hourly']['Hourly Rate'])
hourly_workers
stats.ttest_1samp(hourly_workers, popmean=30).pvalue

# Our p value is very low, so we can reject the hypothesis that the mean of the hourly wage of all hourly workers is 30.
```

We are also curious about salaries in the police force. The chief of police in Chicago claimed in a press briefing that salaries this year are higher than last year's mean of $86000/year a year for all salaried employees. Test this one sided hypothesis using a 95% confidence interval.

Hint: A one tailed test has a p-value that is half of the two tailed p-value. If our hypothesis is greater than, then to reject, the test statistic must also be positive.

```python
police_salaries = salaries.loc[(salaries['Salary or Hourly']=='Salary') & (salaries['Department']=='POLICE')]
police_salaries
stats.ttest_1samp(police_salaries["Annual Salary"], popmean=86000, alternative='less').pvalue
# we want to find out if our given population has a mean annual salary > 86 000 
# our null hypothesis for the ttest is that the sample has a LOWER mean than the theoretical sample (which has a mean of 86000)
# since we want to find the opposite 
# so in the function, we state "alternative=less" because that would be the alternative result if our null hypothesis was correct
# the p value is of 0.99, so we can't reject the hypothesis that the mean salary is lower than last year's. 
```

# Challenge 3: To practice - Constructing Confidence Intervals

While testing our hypothesis is a great way to gather empirical evidence for accepting or rejecting the hypothesis, another way to gather evidence is by creating a confidence interval. A confidence interval gives us information about the true mean of the population. So for a 95% confidence interval, we are 95% sure that the mean of the population is within the confidence interval. 
).

To read more about confidence intervals, click [here](https://en.wikipedia.org/wiki/Confidence_interval).


In the cell below, we will construct a 95% confidence interval for the mean hourly wage of all hourly workers. 

The confidence interval is computed in SciPy using the `t.interval` function. You can read more about this function [here](https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.t.html).

To compute the confidence interval of the hourly wage, use the 0.95 for the confidence level, number of rows - 1 for degrees of freedom, the mean of the sample for the location parameter and the standard error for the scale. The standard error can be computed using [this](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.sem.html) function in SciPy.

```python
hourly_workers 
```

```python
mean_wage = np.mean(hourly_workers)
mean_wage
```

```python
standard_error_mean= stats.sem(hourly_workers) 
standard_error_mean
```

```python
stats.t.interval(0.95, len(hourly_workers)-1, loc=mean_wage, scale=standard_error_mean) 

# for a series named a = interval(0.95, len(a)-1, loc=np.mean(a), scale=st.sem(a))
#syntax : interval (confidence range, degrees of freedom, location parameter - mean of sample, scale - standard error)
```

Now construct the 95% confidence interval for all salaried employeed in the police in the cell below.

```python
salaries_in_police= list(police_salaries["Annual Salary"])
```

```python
stats.t.interval(0.95, len(salaries_in_police)-1, loc=np.mean(salaries_in_police), scale=stats.sem(salaries_in_police)) 
```

# Bonus Challenge - Hypothesis Tests of Proportions

Another type of one sample test is a hypothesis test of proportions. In this test, we examine whether the proportion of a group in our sample is significantly different than a fraction. 

You can read more about one sample proportion tests [here](http://sphweb.bumc.bu.edu/otlt/MPH-Modules/BS/SAS/SAS6-CategoricalData/SAS6-CategoricalData2.html).

In the cell below, use the `proportions_ztest` function from `statsmodels` to perform a hypothesis test that will determine whether the number of hourly workers in the City of Chicago is significantly different from 25% at the 95% confidence level.

```python
#conda install -c conda-forge statsmodels
```

```python
import statsmodels
```

```python
from statsmodels.stats.proportion import proportions_ztest
```

```python
cnt= list(salaries["Salary or Hourly"].value_counts())[1]
cnt
```

```python
no= sum(list(salaries["Salary or Hourly"].value_counts()))
no
```

```python
cnt/no
```

```python
#statsmodels.stats.proportion.proportions_ztest(count, nobs, value=None, alternative='two-sided', prop_var=False)
proportions_ztest(count=cnt, nobs=no, value=0.25, alternative='two-sided')
# noter que le niveau de confiance à 95% est hard-codé dans la fonction 
# donc on spécifie : count as count of "successes" (ici, salary=hourly)
# nobs : number of observations (ici, tous les individus == sample size)
# value : la proportion à laquelle on cherche à comparer notre sample - ici, 25% 
# sachant que si on fait cnt/no == 0.24 
# donc on cherche une p-value pour savoir si on a des chances de trouver des valeurs plus extremes que 0.24 
# la fonction va prendre la distribution cnt/no, centrer réduire autour de 0, puis calculer la p-value. 
```

```python
from scipy import stats
p = cnt/no
p0 = 0.25
2 * stats.norm.cdf(p, loc=p0, scale=np.sqrt(p*(1-p)/no))
# fait à la main par Thomas : avec p (proportion observée), loc = centrer sur proportion de la pop de référence, 
# scale = variance de cnt/no si distribution normale 
# cdf = tout ce qu'il y a en dessous
# 2 fois parce qu'on cherche ce qu'il y a en dessous et au-dessus (de l'autre côté de la courbe)
```

```python
z = (p - p0) / np.sqrt(p*(1-p)/no)
# avec ceci, on crée le z qui centre et réduit 
```

```python
2 * stats.norm.cdf(z, loc=0, scale=1)
```
