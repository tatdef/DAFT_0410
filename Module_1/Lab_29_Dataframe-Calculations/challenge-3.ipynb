```python
orders.drop(columns=["total_spent"], inplace=True)
orders
```

```python
be=pd.DataFrame(orders.groupby("CustomerID")["amount_spent"].sum())
be.head()
```

```python
orders2=orders.merge(be, on="CustomerID")
orders2
```

```python
list_quant=be.quantile([0, 0.25, 0.75, 0.95])
list_quart= list(list_quant["amount_spent"])
list_quart
#The first group is **VIP Customers** whose **aggregated expenses** at your global chain stores are **above the 95th percentile** (aka. 0.95 quantile). 
#The second group is **Preferred Customers** whose **aggregated expenses** are **between the 75th and 95th percentile**.
```

```python
orders2["Label"]="None"
orders2.loc[(orders2["amount_spent_y"]) > list_quart[2], "Label"] = "Preferred"
orders2.loc[(orders2["amount_spent_y"]) > list_quart[3], "Label"] = "VIP"
orders2 
```
